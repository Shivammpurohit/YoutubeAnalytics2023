## Correlation Between Each Column 


```python
# List of numeric columns for which you want to calculate correlation
numeric_columns = ['subscribers', 'video views','uploads','video_views_rank','country_rank','channel_type_rank','video_views_for_the_last_30_days','lowest_monthly_earnings','highest_monthly_earnings','lowest_yearly_earnings','highest_yearly_earnings','subscribers_for_last_30_days']

# Select only the numeric columns from the DataFrame
numeric_data = data[numeric_columns]

# Calculate correlation
correlation = numeric_data.corr()

# Print the correlation matrix
print(correlation)
```

## ------ Correlation Graphs of Each Column with ROws

```python
numeric_columns = ['subscribers', 'video views', 'uploads', 'video_views_rank', 'country_rank', 'channel_type_rank', 'video_views_for_the_last_30_days', 'lowest_monthly_earnings', 'highest_monthly_earnings', 'lowest_yearly_earnings', 'highest_yearly_earnings', 'subscribers_for_last_30_days']

# Select only the numeric columns from the DataFrame
numeric_data = data[numeric_columns]

# Calculate correlation
correlation = numeric_data.corr()

# Create separate scatter plots for each correlation
for column in numeric_columns:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=correlation[column], y=correlation.index, color='Blue')
    plt.title(f'Correlation with {column}')
    plt.tight_layout()
    plt.show()
```

## ----- Heatmap Of The Combined Data

```python
plt.figure(figsize=(15,7))
sns.heatmap(correlation ,vmax=1,annot=True,linewidths=.5)
plt.xticks(rotation=45,horizontalalignment='right')
plt.title('Correlation Heatmap')
plt.show()
```

##### ABOVE HERE WE CAN OBSERVE THAT THE MONTHLY EARNINGS MAJORLY DEPEND UPON THE VIDEO VIEWS AND SUBSCRIBERS

## No. Of Videos in Each Category

```python
cate_counts = data['category'].value_counts()
cate_ordered = cate_counts.index.tolist()

plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='category', palette='viridis',order=cate_ordered)
plt.xticks(rotation=90)
plt.title('Video Categories Count')
plt.ylabel('Number of Videos')
plt.xlabel('Category of Videos')
plt.show()
#Number of videos of each category in top 1000 channel list
```

# Piechart for Videos Category Overall Accumulation

```python
plt.figure(figsize=(8, 8))
plt.pie(cate_counts, labels=cate_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
textprops={'rotation': 120 , 'va': 'center', 'ha': 'left', 'fontsize': 15}
plt.title('Video Categories Distribution')
plt.legend(cate_counts.index, title='Categories', loc='upper right', bbox_to_anchor=(2.3,1))
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


plt.show()

#It is the same chart just in Piechart representation
```

# Channels With Their Country Of Origin

```python
# Calculate the count of each channel
country_counts = data['Country'].value_counts()

# Get the channels in descending order of count
country_ordered = country_counts.index.tolist()

plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Country', palette='plasma',order=country_ordered)
plt.xticks(rotation=90)
plt.title('Channels and Their Country of Origin')
plt.xlabel('Country Of Origin')
plt.ylabel('Count')


plt.show()
```

```python
sample=data.sample(50)
```

```python
sample
```

```python
# Group the data by category and calculate the sum of earnings in the last 30 days for each category
earnings_by_youtuber = sample.groupby('Youtuber')['video_views_for_the_last_30_days'].sum()

# Plot a bar plot
plt.figure(figsize=(10, 6))
earnings_by_youtuber.sort_values(ascending=False).plot(kind='bar', color='purple')
plt.xlabel('Youtuber')
plt.ylabel('Views in Last 30 Days')
plt.title('Total Views in Last 30 Days for each Youtuber')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Total No. Of Uploads In 1 Month Of Top 50 Youtubers
top_50=data.head(50)
top_50.reset_index(drop=True, inplace=True)
top_50
```

```python
top_50_sorted = top_50.sort_values(by='uploads', ascending=False)

plt.figure(figsize=(12, 6))
sns.lineplot(data=top_50_sorted, x='category', y='uploads', marker='o')
plt.xticks(rotation=90)
plt.title('Number of Uploads by Category')
plt.xlabel('Category')
plt.ylabel('Number of Uploads')
plt.tight_layout()

plt.show()
```

```python
# Select the columns you want to plot
columns_to_plot = ['video_views_for_the_last_30_days', 'highest_monthly_earnings']

# Create the pair plot
sns.pairplot(top_50[columns_to_plot])
plt.tight_layout()
# Show the plot
plt.show()
```

#Area Graph of Youtubers with their respective Subscribers

```python
plt.figure(figsize=(10, 6))
data_sorted = top_50.sort_values(by='subscribers', ascending=False)
plt.fill_between(data_sorted['Youtuber'], data_sorted['subscribers'], color='skyblue', alpha=0.4)
plt.plot(data_sorted['Youtuber'], data_sorted['subscribers'], color='blue', marker='o')
plt.xticks(rotation=90)
plt.xlabel('Youtuber')
plt.ylabel('Subscribers')
plt.title('Subscribers of Youtubers (Descending Order)')
plt.tight_layout()
plt.show()
```
